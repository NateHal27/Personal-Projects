##### Six columns have null values: Director, Cast, Country, Date added, Rating, and Duration

#### Too many rows contain null values to drop them all without affecting the data

```python
# replace NaN values in Director with Unknown by using the fillna() method
netflix["director"].fillna("Unknown", inplace = True)
 
# replace NaN values in Cast with Unknown
netflix["cast"].fillna("Unknown", inplace = True)

# replace NaN values in Country with Unknown
netflix["country"].fillna("Unknown", inplace = True)

netflix
```

#### Lets split date added to Months and Years

```python
# convert to datetime using pd.to_datetime
netflix["date_added"] = netflix["date_added"].apply(pd.to_datetime)
netflix.info()
```

```python
# extract the day, month, and year components
netflix['day_added'] = netflix['date_added'].dt.day
netflix['month_added'] = netflix['date_added'].dt.month_name()
netflix['year_added'] = netflix['date_added'].dt.year
```

```python
# View the change
netflix.head(10)
```

```python
# drop unnecessary columns: date_added, and day_added
netflix.drop(columns=['date_added','day_added'], inplace=True, axis=1)
```

### DATA ANALYSIS

#### Q1. What is the Proportion of Movies to Tv shows on Netflix?

```python
# Group type by the count of the show ID
netflix_type = netflix.groupby('type')['show_id'].count()
print(netflix_type)
```

```python
# plot a pie chart
plt.figure(figsize=(8,8))

plt.pie(netflix_type, labels =netflix_type.index, autopct='%1.1f%%')
plt.title('The Proportion of Movies to Tv shows on Netflix')
plt.axis('equal')
plt.show()
```

#### Q2A. How many movies were added each year?

```python
# Separate movies and tv shows
netflix_movies = netflix[netflix['type'] == 'Movie']
netflix_tv = netflix[netflix['type'] == 'TV Show']
```

```python
# Group year added by the count of the show ID
movie_added = netflix_movies.groupby('year_added')['show_id'].count()
print(movie_added)
```

```python
# plot a bar chart to visualize the data
plt.figure(figsize=(15,8))

movie_added.plot(kind='bar',color = 'red')
plt.xlabel('Year added to Netflix')
plt.ylabel('Amount')
plt.title('Addition of New movies to Netflix per Year')
plt.show()
```

#### Q2B. How many tv shows were added each year?

```python
# Group year added by the count of the show ID
tv_added = netflix_tv.groupby('year_added')['show_id'].count()
print(tv_added)
```

```python
# plot a bar chart to visualize the data
plt.figure(figsize=(15,8))

tv_added.plot(kind='bar',color = 'orange')
plt.xlabel('Year added to Netflix')
plt.ylabel('Amount')
plt.title('Addition of New TV Shows to Netflix per Year')
plt.show()
```

#### Q4. What is the average duration of a movie/tv show, the longest movie/tv show and the total amount of time to watch all movies/tv shows?

```python
# Calculate the mean of the duration of movies on netflix
average_movie = netflix_movies['duration_value'].mean()

# Use round syntax to reduce decimal points and print results
print('The average duration of a movie on Netflix is', round(average_movie,2), 'minutes')
```

```python
# Calculate the max of the duration of movies on netflix
max_movie = netflix_movies['duration_value'].max()

# print results
print('The highest duration of a movie on Netflix is', max_movie, 'minutes')
```

```python
# Calculate the sum of the duration of movies on netflix
total_movie = netflix_movies['duration_value'].sum()

# print results
print('The total amount of time to watch all movies on Netflix is', total_movie, 'minutes')
```

```python
# Calculate the mean of the seasons of tv shows on netflix
average_tv = netflix_tv['duration_value'].mean()

# Use round syntax to reduce decimal points and print results
print('The average number of seasons of a Tv show on Netflix is', round(average_tv,2), 'seasons')
```

```python
# Calculate the max of the seasons of tv shows on netflix
max_tv = netflix_tv['duration_value'].max()

# print results
print('The highest number of seasons of a Tv show on Netflix is', max_tv, 'seasons')
```

```python
# Calculate the sum of the seasons of tv shows on netflix
total_tv = netflix_tv['duration_value'].sum()

# print results
print('The total number of seasons of a Tv show on Netflix is', total_tv, 'seasons')
```

#### Q8. Distribution of Mature content on Netflix

```python
# What are the ratings for movies on Netflix?
print(netflix_movies['rating'].unique())
```

```python
# What are the ratings for TV shows on Netflix?
print(netflix_tv['rating'].unique())
```

```python
# Arrange them in order of maturity using a list
mature_mov = ['NR','UR','G','TV-G','TV-Y','TV-Y7','TV-Y7-FV','PG','TV-PG','PG-13','TV-14','R','NC-17','TV-MA']
mature_tv = ['NR','TV-G','TV-Y','TV-Y7','TV-Y7-FV','TV-PG','TV-14','R','TV-MA']
```

```python
movie_rating = netflix_movies['rating'].value_counts()[mature_mov]
tv_rating = netflix_tv['rating'].value_counts()[mature_tv]
```

```python
# plot a pie chart to visualize the data
plt.figure(figsize=(8,8))

plt.pie(movie_rating, labels = movie_rating.index, autopct='%1.1f%%')
plt.title('Distribution of Mature Movies on Netflix')
plt.axis('equal')
plt.legend(bbox_to_anchor=(1.05, 1), loc ='lower right')
plt.show()
```

```python
# plot a pie chart to visualize the data
plt.figure(figsize=(8,8))

plt.pie(tv_rating, labels = tv_rating.index, autopct='%1.1f%%')
plt.title('Distribution of Mature TV Shows on Netflix')
plt.axis('equal')
plt.legend(bbox_to_anchor=(1.05, 1), loc ='lower right')
plt.show()
```
